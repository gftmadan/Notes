```python
import torch as th
import cv2
import numpy as np
import shutil
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def create_yolo_file_structure(Main_LOC,Sub_DIR):
    loc = os.path.join(Main_LOC,Sub_DIR)
    if os.path.exists(loc) == False:
        os.mkdir(loc)
    images_loc = os.path.join(loc,"images")
    labels_loc = os.path.join(loc,"labels")
    if os.path.exists(images_loc) == False:
        os.mkdir(images_loc)
    if os.path.exists(labels_loc) == False:
        os.mkdir(labels_loc)
    return images_loc,labels_loc
```

```python
def pascal_voc_to_yolo(x1, y1, x2, y2, image_w, image_h):
    return [((x2 + x1)/(2*image_w)), ((y2 + y1)/(2*image_h)), (x2 - x1)/image_w, (y2 - y1)/image_h]
```

```python
ANIMAL_ORG_TRAIN_DIR = r"E:\PyImage_ComputerVision\ObjectDetectionYolo\WorkingData\InputData\AnimalDatasetOriginal\train"
ANIMAL_ORG_TEST_DIR = r"E:\PyImage_ComputerVision\ObjectDetectionYolo\WorkingData\InputData\AnimalDatasetOriginal\test"

OUTPUT_DATA_DIR = "E:\PyImage_ComputerVision\ObjectDetectionYolo\WorkingData\InputData\AnimalDatasetYolo"
count = 0
class_idx = dict()
class_count = 0
TRAIN_IMAGES_LOC,TRAIN_LABELS_LOC = create_yolo_file_structure(OUTPUT_DATA_DIR,"train")
for folder in os.listdir(ANIMAL_ORG_TRAIN_DIR):
    folder_loc = os.path.join(ANIMAL_ORG_TRAIN_DIR,folder)
    if os.path.isfile(folder_loc) == True:
        continue
    else:
        for file in os.listdir(folder_loc):
            file_loc = os.path.join(folder_loc,file)
            if os.path.isfile(file_loc) == True:
                file_name = os.path.splitext(file)[0]
                label_file_loc = os.path.join(folder_loc,"Label",file_name+".txt")
                output_label_loc = os.path.join(TRAIN_LABELS_LOC,f"{count}.txt")
                output_image_loc = os.path.join(TRAIN_IMAGES_LOC,f"{count}.png")
                count +=1
                img = cv2.imread(file_loc)
                ht,wt,_ = img.shape
                f1 = open(label_file_loc,"r")
                print_buffer = []
                for line in f1.readlines():
                    *cls,x1,y1,x2,y2 = line.split()
                    cls = "".join(cls)
                    if cls not in class_idx:
                        class_idx[cls] = class_count
                        cls = class_count
                        class_count += 1
                    else:
                        cls = class_idx[cls]
                    cx,cy,w,h = pascal_voc_to_yolo(float(x1),float(y1),float(x2),float(y2),wt,ht)
                    yolo_str = f"{cls} {cx} {cy} {w} {h}"
                    print_buffer.append(yolo_str)
                f1.close()
                f2 = open(output_label_loc,"w")
                f2.write("\n".join(print_buffer))
                f2.close()
                cv2.imwrite(output_image_loc,img)


count = 0
TRAIN_IMAGES_LOC,TRAIN_LABELS_LOC = create_yolo_file_structure(OUTPUT_DATA_DIR,"test")
for folder in os.listdir(ANIMAL_ORG_TEST_DIR):
    folder_loc = os.path.join(ANIMAL_ORG_TEST_DIR,folder)
    if os.path.isfile(folder_loc) == True:
        continue
    else:
        for file in os.listdir(folder_loc):
            file_loc = os.path.join(folder_loc,file)
            if os.path.isfile(file_loc) == True:
                file_name = os.path.splitext(file)[0]
                label_file_loc = os.path.join(folder_loc,"Label",file_name+".txt")
                output_label_loc = os.path.join(TRAIN_LABELS_LOC,f"{count}.txt")
                output_image_loc = os.path.join(TRAIN_IMAGES_LOC,f"{count}.png")
                count +=1
                img = cv2.imread(file_loc)
                ht,wt,_ = img.shape
                f1 = open(label_file_loc,"r")
                print_buffer = []
                for line in f1.readlines():
                    *cls,x1,y1,x2,y2 = line.split()
                    cls = "".join(cls)
                    if cls not in class_idx:
                        class_idx[cls] = class_count
                        cls = class_count
                        class_count += 1
                    else:
                        cls = class_idx[cls]
                    cx,cy,w,h = pascal_voc_to_yolo(float(x1),float(y1),float(x2),float(y2),wt,ht)
                    yolo_str = f"{cls} {cx} {cy} {w} {h}"
                    print_buffer.append(yolo_str)
                f1.close()
                f2 = open(output_label_loc,"w")
                f2.write("\n".join(print_buffer))
                f2.close()
                cv2.imwrite(output_image_loc,img)     
```
