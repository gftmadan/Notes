# Sorting Pandas Object Using:- `sort_values()`

## 1) Sorting the DataFrame by Single Column(age) values 

```python
fg1 = df.sort_values(by=["age"])
fg1
```

```python
fg1.iloc[1000:5000]
```

## 2)Sorting the DataFrame According to Single Column(age) values in Descending Order

```python
fg2 = df.sort_values(by=["age"],ascending=False)
fg2
```

```python
fg2.iloc[30000:35000]
```

## 3)Sorting The DataFrame According to Multiple Columns(age,blood_glucose_level)

```python
fg3 = df.sort_values(by=["age","blood_glucose_level"])
fg3
```

```python
fg3.iloc[10000:10500]
```

```python
fg4 = df.sort_values(by=["blood_glucose_level","age"])
fg4
```

```python
fg3.iloc[20000:25000]
```

## 4)Sorting the DataFrame According to the Multiple Columns Sorting way

That means in which way either Ascending or Descending the columns should the affect the sorting!

```python
fg5 = df.sort_values(by=["age","bmi"],ascending=[False,True])
fg5
```

## 5) Using key to manipulate the Sorting of the DataFrame

### a)Using Lambda Function for sorting in advance way

```python
fg6 = df.sort_values(by=["smoking_history"],key=lambda x:x.apply(str).apply(len))
fg6
```

```python
fg7 = df.sort_values(by=["smoking_history"],key=lambda x:x.apply(str).apply(len),ascending=False)
fg7
```

## 6)Sorting the DataFrame according to the Column containing NaN Values(Managing Them)

```python
df3 =df.copy()
df3.loc[[0,1,2],"bmi"] = np.NaN
```

```python
df3
```

```python
fg12 = df3.sort_values(by=["bmi"])
fg12
```

```python
fg13 = df3.sort_values(by=["bmi"],ascending=False)
fg13
```

```python
fg14 = df3.sort_values(by=["bmi"],na_position="first")
fg14
```

```python
fg15 = df3.sort_values(by=["bmi"],na_position="first",ascending=False)
fg15
```

# Sorting Pandas Objects Using:- `sort_index()` function

```python
df
```

```python
fg1 = df.sort_index(axis=1,key=lambda x:[len(fg) for fg in x])
fg1
```

```python
fg2 = df.sort_index(axis=1,key=lambda x:[len(fg) for fg in x],ascending=False)
fg2
```

```python
fg3 = df.sort_index(axis=0,ascending=False)
fg3
```

```python
fgdf =df.copy()
fgdf.sort_index(axis=1,key=lambda x:[len(fg) for fg in x],inplace=True,ignore_index=True)
fgdf
```

```python
fgdf =df.copy()
fgdf.sort_index(axis=0,ascending=False,inplace=True,ignore_index=True)
fgdf
```
