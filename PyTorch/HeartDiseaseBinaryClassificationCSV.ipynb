```python
from torch import nn
import pandas as pd
import numpy as np
import torch as th
```

```python
df_data = pd.read_csv(r"D:\Machine Learning\Sample Project1\Notes Notebooks\Working_Data\heart-disease.csv")
```

```python
df_data
```

```python
X_data = df_data.drop("target",axis=1)
```

```python
X_data
```

```python
Y_data = df_data["target"]
```

```python
Y_data
```

```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer

OHE = OneHotEncoder(sparse_output=False)
SS = StandardScaler()

CTF = ColumnTransformer([("OHE",OHE,["sex","cp","fbs","restecg","exang","slope","ca","thal"]),("SS",SS,["age","trestbps","chol","thalach","oldpeak"])],remainder="passthrough")
CTF.set_output(transform="pandas")

X_final_data = CTF.fit_transform(X_data)
```

```python
X_final_data
```

```python
Y_data
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X_final_data,Y_data,test_size=0.2,random_state=42)
```

```python
X_train
```

```python
Y_train
```

```python
X_test
```

```python
Y_test
```

```python
X_train_tensor = th.tensor(X_train.values).type(th.float32)
X_test_tensor = th.tensor(X_test.values).type(th.float32)
Y_train_tensor = th.tensor(Y_train.values).type(th.float32)
Y_test_tensor = th.tensor(Y_test.values).type(th.float32)
```

```python
DEVICE = "cuda" if th.cuda.is_available() else "cpu"
```

```python
DEVICE
```

```python
model_V0 = nn.Sequential(
                            nn.Linear(in_features=30,out_features=16),
                            nn.ReLU(),
                            nn.Dropout(p=0.3),
                            nn.Linear(in_features=16,out_features=8),
                            nn.ReLU(),
                            nn.Dropout(p=0.3),
                            nn.Linear(in_features=8,out_features=1)
)
```

```python
model_V0.to(DEVICE)
```

```python
model_V0
```

```python
trainable_params = sum(
    p.numel() for p in model_V0.parameters()
)
trainable_params
```

```python
loss_fn = nn.BCEWithLogitsLoss()

optimizer = th.optim.Adam(model_V0.parameters(),
                          lr=0.001)
```

```python
X_Dl_train = th.utils.data.DataLoader(X_train_tensor,batch_size=32,shuffle=False)
Y_Dl_train = th.utils.data.DataLoader(Y_train_tensor,batch_size=32,shuffle=False)
X_Dl_test = th.utils.data.DataLoader(X_test_tensor,batch_size=32,shuffle=False)
Y_Dl_test = th.utils.data.DataLoader(Y_test_tensor,batch_size=32,shuffle=False)
```

```python
def accuracy(y_true,y_preds):
    correct = th.eq(y_preds,y_true).sum().item()
    acc = (correct/len(y_preds))*100
    return acc
```

```python
def evaluate(Model,X_DL,Y_DL):
    total_accuracy = 0.0
    Model.to(DEVICE)
    Model.eval()
    with th.no_grad():
        for X,Y in zip(X_DL,Y_DL):
            X = X.to(DEVICE)
            Y = Y.to(DEVICE)
            y_logits = Model(X).squeeze()
            y_preds = th.round(th.sigmoid(y_logits))
            acc = accuracy(Y,y_preds)
            total_accuracy += acc
    total_accuracy /= len(X_DL)
    print(f"The Evaluation Accuracy = {total_accuracy:.2f}")
```

```python
EPOCHS = 700
BATCH_SIZE = 32
TRAIN_DL_LEN = len(X_Dl_train)
TEST_DL_LEN = len(X_Dl_test)
model_V0.to(DEVICE)



for epoch in range(EPOCHS):
    
    
    model_V0.train()
    epoch_train_loss = 0.0
    epoch_train_accuracy = 0.0
    for X,Y in zip(X_Dl_train,Y_Dl_train):
        X = X.to(DEVICE)
        Y = Y.to(DEVICE)
        
        y_logits = model_V0(X).squeeze()
        
        loss = loss_fn(y_logits,Y)
        y_preds = th.round(th.sigmoid(y_logits))
        accr = accuracy(Y,y_preds)
        
        epoch_train_loss += loss
        epoch_train_accuracy += accr
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    
    
    model_V0.eval()
    with th.no_grad():
        epoch_test_loss = 0.0
        epoch_test_accuracy = 0.0
        for X,Y in zip(X_Dl_test,Y_Dl_test):
            X = X.to(DEVICE)
            Y = Y.to(DEVICE)
            y_logits = model_V0(X).squeeze()
            
            loss = loss_fn(y_logits,Y)
            y_preds = th.round(th.sigmoid(y_logits))
            accr = accuracy(Y,y_preds)
            
            epoch_test_loss += loss
            epoch_test_accuracy += accr
    
    
    epoch_train_loss /= TRAIN_DL_LEN
    epoch_train_accuracy /= TRAIN_DL_LEN
    epoch_test_loss /= TEST_DL_LEN
    epoch_test_accuracy /= TEST_DL_LEN
    
    
    print(f"Epoch {epoch} :- |Train Loss:- {epoch_train_loss:.2f} <--------> Train Accuracy:- {epoch_train_accuracy:.2f}|++++++++|Test Loss:- {epoch_test_loss:.2f} <--------> Test Accuracy :- {epoch_test_accuracy:.2f}|")
```

```python
evaluate(model_V0,X_Dl_test,Y_Dl_test)
```
