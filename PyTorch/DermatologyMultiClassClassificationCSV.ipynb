```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer

OHE = OneHotEncoder(sparse_output=False)
SS = StandardScaler()

CTF = ColumnTransformer([("OHE",OHE,OHE_columns),("SS",SS,["age"])],remainder="passthrough")
CTF.set_output(transform="pandas")

X_final_data = CTF.fit_transform(X_data)
```

```python
OHE_Y = OneHotEncoder(sparse_output=False)

Y_final_data = OHE_Y.fit_transform(Y_data.values.reshape(-1,1))

Y_final_data
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X_final_data,Y_final_data,test_size=0.2,random_state=42)
```

```python
X_tensor_train = th.Tensor(X_train.values).type(th.float32)
X_tensor_test = th.Tensor(X_test.values).type(th.float32)
```

```python
X_DL_train = th.utils.data.DataLoader(X_tensor_train,batch_size=32,shuffle=False)
X_DL_test = th.utils.data.DataLoader(X_tensor_test,batch_size=32,shuffle=False)
Y_DL_train = th.utils.data.DataLoader(Y_train,batch_size=32,shuffle=False)
Y_DL_test = th.utils.data.DataLoader(Y_test,batch_size=32,shuffle=False)
```

```python
DEVICE ="cuda" if th.cuda.is_available() else "cpu"
```

```python
DEVICE
```

```python
model_V0 = nn.Sequential(
                            nn.Linear(in_features=130,out_features=256),
                            nn.ReLU(),
                            nn.Dropout(0.3),
                            nn.Linear(in_features=256,out_features=64),
                            nn.ReLU(),
                            nn.Dropout(0.3),
                            nn.Linear(in_features=64,out_features=16),
                            nn.ReLU(),
                            nn.Dropout(0.3),
                            nn.Linear(in_features=16,out_features=6)
)
```

```python
model_V0
```

```python
trainable_params = sum(
    p.numel() for p in model_V0.parameters()
)
trainable_params
```

```python
loss_fn = nn.CrossEntropyLoss()

optimizer = th.optim.Adam(model_V0.parameters(),
                          lr=0.001)
```

```python
def accuracy(y_preds,y_true):
    correct = th.eq(y_preds,y_true).sum().item()
    acc = (correct/len(y_preds))*100
    return acc
```

```python
def evaluate(model,X_test,Y_test):
    model.eval()
    model.to(DEVICE)
    TEST_DL_LEN = len(X_test)
    total_acc = 0.0
    with th.no_grad():
        for i,(X,Y) in enumerate(zip(X_test,Y_test)):
            X = X.to(DEVICE)
            Y = Y.to(DEVICE)
            y_logits = model(X).squeeze()
            y_preds = th.argmax(th.softmax(y_logits,1),1)
            y_true = th.argmax(Y,1)
            acc = accuracy(y_preds,y_true)
            total_acc += acc
            print(f"Batch {i}   Accuracy={acc}")
    
    total_acc /= TEST_DL_LEN
    print(f"The Average Model Accuracy is {total_acc:.3f}")
```

```python
EPOCHS = 100
BATCH_SIZE = 32
TRAIN_DL_LEN = len(X_DL_train)
TEST_DL_LEN = len(X_DL_test)


model_V0.to(DEVICE)
for epoch in range(EPOCHS):
    model_V0.train()
    epoch_train_loss = 0.0
    epoch_train_accuracy = 0.0
    for X,Y in zip(X_DL_train,Y_DL_train):
        X = X.to(DEVICE)
        Y = Y.to(DEVICE)
        
        y_logits = model_V0(X).squeeze()
        y_preds = th.argmax(th.softmax(y_logits,1),1)
        y_true = th.argmax(Y,1)
        
        loss = loss_fn(y_logits,Y)
        acc = accuracy(y_preds,y_true)
        
        epoch_train_loss += loss
        epoch_train_accuracy += acc
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    
    
    
    model_V0.eval()
    epoch_test_loss = 0.0
    epoch_test_accuracy = 0.0
    with th.no_grad():
        for X,Y in zip(X_DL_test,Y_DL_test):
            X = X.to(DEVICE)
            Y = Y.to(DEVICE)
            
            y_logits = model_V0(X).squeeze()
            y_preds = th.argmax(th.softmax(y_logits,1),1)
            y_true = th.argmax(Y,1)
            
            loss = loss_fn(y_logits,Y)
            accr = accuracy(y_preds,y_true)
            
            epoch_test_loss += loss
            epoch_test_accuracy += accr
            
    
    epoch_train_accuracy /= TRAIN_DL_LEN
    epoch_train_loss /= TRAIN_DL_LEN
    epoch_test_accuracy /= TEST_DL_LEN
    epoch_test_loss /= TEST_DL_LEN
    
    
    print(f"Epoch {epoch} :- |Train Loss:- {epoch_train_loss:.2f} <--------> Train Accuracy:- {epoch_train_accuracy:.2f}|++++++++|Test Loss:- {epoch_test_loss:.2f} <--------> Test Accuracy :- {epoch_test_accuracy:.2f}|")
```

```python
evaluate(model_V0,X_DL_test,Y_DL_test)
```
