# Image Histograms

### An Image Histogram represents the Distribution of Image pixel Intensities(Colored or GrayScale) in an Image,

### -->Its an Visualization Tool that gives a HIgh Level Intution of Intensity(Pixels Value) Distribution.

### -->When Plotting the Histogram ,the X-Axis Servers as `Bins`. If we Have 256 Bins then we are Effectively Counting the Number of Times each Pixel Value Occurs.

### -->If we have 2 `Bins` then we are Counting the Number of Times a Pixel is in Range of [1,128] ,[129,255].

### -----------------------------------------------------------------------------------------------------

```python
def plot_hist(hist):
    hist = hist/hist.sum()
    plt.figure(figsize=(7,7))
    plt.xlabel("Bins")
    plt.ylabel("% of Pixels")
    plt.plot(hist)
```

### In the Above Cells we Created a method Which Helps us in Plotting the Histogram using MatPlotlib Library.

### -----------------------------------------------------------------------------------------------------

### -->We use `cv2.calcHist()` Method to compute our  image Histograms

## The Function Defination is `cv2.calcHist(images,channels,mask,histsize,ranges)` :-

### < images >:- It takes the List of Numpy Array of the Images of which we want to Compute Image Histogram.But If we have a Single Input Image we have to wrap it up as List.

### < channels >:-It takes the List of Indexs.These Indexs specifies the Channel we want to Compute the Histogram for.To compute a Histogram of GrayScale Image it would be [0] , and for B,G,R Channel it would be [0,1,2].

### < mask >:-If a Mask is passed then the Image Histogram will be only be Computed for the Masked Pixels.If None then it will be Computed for the Whole Image.

### < histsize >:-It takes the List of the Number of Bins we want to use for Each Channel for the Computation of Image Histograms.

### < ranges >:-It takes the List of Possible Range Value for Each Channel.It uses LowerBound and UpperBound-1.Hence we use (0,256) values for RGB Color Channel which means Consider Range from 0-255.

```python
def plot_color_hist(image):
    sp = cv2.split(image)
    clrs = ["b","g","r"]
    for ch,clr in zip(sp,clrs):
        plt.figure(figsize=(5,5))
        plt.ylabel("% of Pixels")
        plt.xlabel("Bins")
        hist = cv2.calcHist([ch],[0],None,[256],[0,256])
        hist = hist/hist.sum()
        plt.plot(hist,color=clr)
```
