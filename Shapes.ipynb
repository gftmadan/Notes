## cv2.arcLenght() Method:-https://docs.opencv.org/4.x/d3/dc0/group__imgproc__shape.html#ga8d26483c636be6b35c3ec6335798a47c

## The perimeter is the distance around the outside of a shape. It is calculated by adding up the lengths of all the sides of the shape. In OpenCV, the perimeter of a contour is calculated by adding up the lengths of all the edges of the contour.
The units of the perimeter returned by the arcLength() function depend on the units of the image that the contour was extracted from. For example, if the image is in pixels, then the perimeter will be in pixels. If the image is in millimeters, then the perimeter will be in millimeters.

```python
def DetectShape(contour):
    per = cv2.arcLength(contour,True)
    shape = None
    apx = cv2.approxPolyDP(contour,0.04 * per,True)
    vertices = apx.shape[0]
    if vertices > 4:
        if vertices == 5:
            shape = "Polygon"
        else:
            shape = "Circle"
    else:
        if vertices == 4:
            (x,y,w,h) = cv2.boundingRect(apx)
            rt = float(w)/float(h)
            shape = "Square" if rt >= 0.95 and rt <= 1.05 else "Rectangle"
        if vertices == 3:
            shape = "Traingle"
    return shape
```

```python
cnts = cv2.findContours(thresh1[1],cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
```

```python
shape_out = img1.copy()
for idx,cnt in enumerate(cnts[0][0:17]):
    shape = DetectShape(cnt)
    M = cv2.moments(cnt)
    print(idx)
    Cx = int(M["m10"] / M["m00"])
    Cy = int(M["m01"] / M["m00"])
    cv2.drawContours(shape_out,[cnt],-1,(0,255,0),2)
    cv2.putText(shape_out,shape,(Cx,Cy),cv2.FONT_HERSHEY_SIMPLEX,0.5,(255,255,255),2)
```

```python
cv2.imshow("Detected Shapes",shape_out)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
