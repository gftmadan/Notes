```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output=False)
clt = ColumnTransformer([("ohe",ohe,["gender","smoking_history"])],remainder="passthrough")
clt.set_output(transform="pandas")
clt.fit_transform(X)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

oheg = OneHotEncoder(sparse_output=False)
ohes = OneHotEncoder(sparse_output=False,min_frequency=10000)

clt = ColumnTransformer([("oheg",oheg,["gender"]),
                         ("ohes",ohes,["smoking_history"])],remainder="passthrough")
clt.set_output(transform="pandas")
clt.fit(X)
clt.transform(X)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

oheg = OneHotEncoder(sparse_output=False)
ohes = OneHotEncoder(sparse_output=False,min_frequency=10000)

clt = ColumnTransformer([("oheg",oheg,[0]),
                         ("ohes",ohes,[4])],remainder="passthrough")
clt.set_output(transform="pandas")
clt.fit(X)
clt.transform(X)
```
