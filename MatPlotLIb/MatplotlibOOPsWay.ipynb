## Creating Sccater plot using OOPS Way

```python
df_diab = pd.read_csv(r"D:\Machine Learning\Sample Project1\Notes Notebooks\Working_Data\diabetes_dataset.csv")
```

```python
df_diab_small = df_diab.sample(n=100)
```

```python
df_diab_small
```

```python
x = df_diab_small["bmi"]
y = df_diab_small["blood_glucose_level"]
s = df_diab_small["age"]
c = df_diab_small["diabetes"]

fig,ax = plt.subplots(nrows=1,ncols=1,figsize=(10,10))

sct = ax.scatter(x,y,s=s,c=c)

ax.minorticks_on()
ax.grid(which="major",axis="both",linestyle="-",linewidth=1.2)
ax.grid(which="minor",axis="both",linestyle="--",linewidth=0.6)

ax.set_xlabel("BMI",labelpad=15.0,fontsize="large")
ax.set_ylabel("Blood Glucose Level",labelpad=15.0,fontsize="large")
ax.set_title("Blood Glucose Level VS BMI",pad=15.0,fontsize="xx-large")

first_legend = plt.legend(*sct.legend_elements(prop="colors"),title="Diabetes",loc="upper right",borderpad=1.1)
ax.legend(*sct.legend_elements(prop="sizes"),title="Age",loc="upper left",borderpad=1.1)
ax.add_artist(first_legend);
```

# Creating Histograms using OOPS WAY with annotations and lines

```python
plt.style.use("ggplot")
x = df_diab["bmi"]
bins=[10,20,30,40,50]

fig,ax  = plt.subplots(nrows=1,ncols=1,figsize=(10,6))

hist = ax.hist(x,bins,color="#FFCCCB",edgecolor="black",linewidth=2,zorder=5)

ax.minorticks_on()
ax.tick_params(axis="x",which="minor",bottom=False)
ax.grid(axis="y",which="major",linestyle="-",linewidth=1.2,color="black")
ax.grid(axis="y",which="minor",linestyle="--",linewidth=0.6,color="black")

ax.set_xticks(bins,bins)
ax.set_xlabel("BMI Range",labelpad=30,fontsize="large")
ax.set_ylabel("Counts of BMI",labelpad=15.0,fontsize="large")
ax.set_title("Histogram of BMI",pad=15.0,fontsize="xx-large")

freqs,bins,fg = hist

bins = 0.5 * np.diff(bins) + bins[:-1]

pers = np.round(freqs/freqs.sum() * 100,1)

for freq,per,bn in zip(freqs,pers,bins):
    
    ax.annotate(f"{str(freq)}",xy=(bn,freq),xytext=(0,2.2),textcoords="offset points",ha="center",zorder=10,fontsize="large",color="red")
    ax.annotate(f"{str(per)}%",xy=(bn,0),xytext=(0,-15),textcoords="offset points",ha="center",zorder=10,fontsize="large")
```
