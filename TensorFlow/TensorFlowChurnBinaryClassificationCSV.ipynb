```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
X = churn_df.drop(["RowNumber","CustomerId","Surname","Exited"],axis=1,inplace=False)
X
```

```python
y = churn_df["Exited"]
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer


ohc = OneHotEncoder(sparse_output=False)
ctf = ColumnTransformer([("ohc",ohc,["Gender","Geography"])],remainder="passthrough")
ctf.set_output(transform="pandas")
X_train1 = ctf.fit_transform(X_train)
X_test1 = ctf.transform(X_test)
```

```python
X_train1
```

```python
X_test1
```

```python
X_train1.info()
```

```python
from sklearn.preprocessing import StandardScaler

scl = StandardScaler()
scl.set_output(transform="pandas")

X_train = scl.fit_transform(X_train1)
X_test = scl.transform(X_test1)
```

```python
X_train
```

```python
X_test
```

```python
model = tf.keras.Sequential()
```

```python
model.add(tf.keras.layers.Dense(64,input_dim=13,activation="relu"))
model.add(tf.keras.layers.Dense(32,activation="relu"))
model.add(tf.keras.layers.Dense(16,activation="sigmoid"))
model.add(tf.keras.layers.Dense(1,activation="sigmoid"))
```

```python
model.summary()
```

```python
model.compile(optimizer="Adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
with tf.device('/GPU:0'):
    model.fit(X_train,y_train,epochs=50,validation_split=0.15)
```

```python
y_log = model.predict(X_test)
y_pred = np.where(y_log > 0.5, 1, 0)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
