```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer

SS = StandardScaler()
OHE = OneHotEncoder(sparse_output=False)

CTF = ColumnTransformer([("OHE",OHE,["sex","smoker","region"]),("SS",SS,["age","bmi","children"])],remainder="passthrough")
CTF.set_output(transform="pandas")

X1 = CTF.fit_transform(X_data)
```

```python
X1
```

```python
from sklearn.model_selection import train_test_split

X1_train,X1_test,Y1_train,Y1_test = train_test_split(X1,Y_data,test_size=0.2)
```

```python
X1_train
```

```python
model_V0 = tf.keras.Sequential([
                                tf.keras.layers.Dense(64,activation="relu"),
                                tf.keras.layers.Dense(32,activation="relu"),
                                tf.keras.layers.Dense(16,activation="relu"),
                                tf.keras.layers.Dense(1,activation="linear")
])
```

```python
model_V0.compile(loss="mse",
                 optimizer=tf.keras.optimizers.Adam(0.01))
```

```python
model_V0.fit(X1_train,Y1_train,epochs=1000,batch_size=32)
```

```python
Y_preds = model_V0.predict(X1_test)
```
