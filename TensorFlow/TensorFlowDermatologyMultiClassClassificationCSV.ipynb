```python
df_data["age"] = df_data["age"].replace("?",np.NaN)
```

```python
df_data.dropna(inplace=True)
```

```python
df_data
```

```python
OHE_columns = list(df_data.columns)
OHE_columns.remove("age")
OHE_columns.remove("class")
```

```python
OHE_columns
```

```python
X_data = df_data.drop("class",axis=1)
X_data["age"] = X_data["age"].astype('float')
X_data
```

```python
X_data["age"].value_counts().index
```

```python
Y_data = df_data["class"]
Y_data
```

```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer

OHE = OneHotEncoder(sparse_output=False)
SS = StandardScaler()

CTF = ColumnTransformer([("OHE",OHE,OHE_columns),("SS",SS,["age"])],remainder="passthrough")
CTF.set_output(transform="pandas")

X_final_data = CTF.fit_transform(X_data)
```

```python
X_final_data
```

```python
OHE_Y = OneHotEncoder(sparse_output=False)

Y_final_data = OHE_Y.fit_transform(Y_data.values.reshape(-1,1))

Y_final_data
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X_final_data,Y_final_data,test_size=0.2,random_state=42)
```

```python
X_train
```

```python
len(Y_train)
```

```python
X_test
```

```python
Y_test.shape
```

```python
model_V0 = tf.keras.Sequential([
                                tf.keras.layers.InputLayer(input_shape=(130,)),
                                tf.keras.layers.Dense(256,activation="relu"),
                                tf.keras.layers.Dropout(0.3),
                                tf.keras.layers.Dense(64,activation="relu"),
                                tf.keras.layers.Dropout(0.3),
                                tf.keras.layers.Dense(16,activation="relu"),
                                tf.keras.layers.Dropout(0.3),
                                tf.keras.layers.Dense(6,activation="softmax")
])
```

```python
model_V0.summary()
```

```python
model_V0.compile(loss=tf.keras.losses.CategoricalCrossentropy(),
                 optimizer=tf.keras.optimizers.Adam(),
                 metrics=["accuracy"])
```

```python
model_V0.fit(X_train,Y_train,validation_data=(X_test,Y_test),validation_batch_size=32,batch_size=32,epochs=100)
```

```python
model_V0.evaluate(X_test,Y_test)
```
