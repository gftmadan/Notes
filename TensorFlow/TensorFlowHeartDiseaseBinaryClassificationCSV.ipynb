```python
X_data = df_data.drop("target",axis=1)
```

```python
X_data
```

```python
Y_data =  df_data["target"]
```

```python
Y_data
```

```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer

OHE = OneHotEncoder(sparse_output=False)
SS = StandardScaler()

CTF = ColumnTransformer([("OHE",OHE,["sex","cp","fbs","restecg","exang","slope","ca","thal"]),("SS",SS,["age","trestbps","chol","thalach","oldpeak"])],remainder="passthrough")
CTF.set_output(transform="pandas")

X_final_data = CTF.fit_transform(X_data)
```

```python
X_final_data
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X_final_data,Y_data,test_size=0.2,random_state=42)
```

```python
X_train
```

```python
Y_train
```

```python
X_test
```

```python
Y_test
```

```python
model_V0 = tf.keras.Sequential([
                                tf.keras.layers.InputLayer(input_shape=(30,)),
                                tf.keras.layers.Dense(16,activation="relu"),
                                tf.keras.layers.Dropout(0.3),
                                tf.keras.layers.Dense(8,activation="relu"),
                                tf.keras.layers.Dropout(0.3),
                                tf.keras.layers.Dense(1,activation="sigmoid")    
])
```

```python
model_V0.compile(loss=tf.keras.losses.BinaryCrossentropy(),
                 optimizer=tf.keras.optimizers.Adam(),
                 metrics=["accuracy"])
```

```python
model_V0.fit(X_train,Y_train,validation_data=(X_test,Y_test),batch_size=32,validation_batch_size=32,epochs=700)
```

```python
model_V0.summary()
```

```python
model_V0.evaluate(X_test,Y_test)
```
