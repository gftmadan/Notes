```python
import cv2
import math
```

```python
def get_video_duration(video):
    frames = video.get(cv2.CAP_PROP_FRAME_COUNT)
    fps = video.get(cv2.CAP_PROP_FPS)
    return frames//fps
```

```python
while(vd1_360.isOpened()):
    ret,frame = vd1_360.read()
    if ret == True:
        cv2.imshow("Frame",frame)
        cv2.waitKey(1)
    else:
        cv2.destroyAllWindows()
        break
```

```python
while(vd1_360.isOpened()):
    frame_num = vd1_360.get(cv2.CAP_PROP_POS_FRAMES)
    mili = vd1_360.get(cv2.CAP_PROP_POS_MSEC)
    avr = vd1_360.get(cv2.CAP_PROP_POS_AVI_RATIO)
    ret,frame = vd1_360.read()
    if ret == True:
        print(f"Frame Number = {frame_num}\nMilliSeconds = {mili}\nAvi Ratio = {avr}")
    else:
        break
```

```python
def reverse_video(video):
    nof = video.get(cv2.CAP_PROP_FRAME_COUNT)
    while(nof>=0):
        video.set(cv2.CAP_PROP_POS_FRAMES,nof-1)
        ret,frame = video.read()
        cv2.imshow("frame",frame)
        cv2.waitKey(3)
        nof -= 1
    cv2.destroyAllWindows()
```

```python
reverse_video(match)
```
